```python
chemtastedb_df = pd.read_excel("../data/ChemTastesDB_database.xlsx",
                               na_values=["*"],
                               keep_default_na=True)
print(chemtastedb_df.info())
chemtastedb_df.head()
```

```python
chemtastedb_df.isnull().sum()
```

```python
taste_count = chemtastedb_df["Class taste"].value_counts(dropna=False)
```

```python
ax = taste_count.plot(kind="bar", title="Taste Distribution")

for i, count in enumerate(taste_count):
    ax.text(i, count, str(count), ha="center", va="bottom")

ax.set_xlabel("Taste Categories")
ax.set_ylabel("Count")
ax.set_xticklabels(taste_count.index.fillna("NA"), rotation=45)

plt.show()
```

```python
pubchemid_na_count = chemtastedb_df["PubChem CID"].isna().sum()
total_entries = len(chemtastedb_df)

labels = ["Valid PubChemCid", "Missing"]
sizes = [total_entries - pubchemid_na_count, pubchemid_na_count]
explode = (0, 0.1)

plt.pie(sizes, explode=explode, labels=labels, startangle=90, autopct='%1.1f%%',
        textprops={'fontsize': 12})
plt.axis('equal')
plt.title('PubChemCID Completeness')
plt.show()
```
